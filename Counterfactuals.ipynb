```python
import polars as pl
```

```python
df1_pl = pl.from_pandas(df1)
df2_pl = pl.from_pandas(df2)
```

```python
df_list = []
for sample_frag in tqdm(df1_pl["bait_frag"].unique()):
    anchor1 = df1_pl.filter(df1_pl["bait_frag"] == sample_frag)
    anchor2 = df2_pl.filter(df2_pl["bait_frag"] == sample_frag)
    if len(anchor2) == 0:
        continue
    df = pl.DataFrame({
        "bait_frag" : sample_frag,
        "other_frag": [
            frag2
            for frag2 in anchor2["other_frag"] 
            if frag2 not in anchor1["other_frag"]
        ]
    })
    df_list.append(df)
```

```python
for sample_frag in tqdm(df1_pl["other_frag"].unique()):
    anchor1 = df1_pl.filter(df1_pl["other_frag"] == sample_frag)
    anchor2 = df2_pl.filter(df2_pl["other_frag"] == sample_frag)
    if len(anchor2) == 0:
        continue
    df = pl.DataFrame({
        "bait_frag": [
            frag2
            for frag2 in anchor2["bait_frag"] 
            if frag2 not in anchor1["bait_frag"]
        ],
        "other_frag" : sample_frag,

    })
    df_list.append(df)
```

```python
counters = pl.concat([df for df in df_list if pl.Null not in df.dtypes]).sample(fraction=1.0, seed=123, shuffle=True)
```

```python
counters.write_csv("data/intermediates/counterfactuals.csv")
```
