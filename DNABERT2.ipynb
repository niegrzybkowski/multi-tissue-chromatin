```python
import torch
from transformers import AutoTokenizer, AutoModel

tokenizer = AutoTokenizer.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True)
model = AutoModel.from_pretrained("zhihan1996/DNABERT-2-117M", trust_remote_code=True).cuda()
```

```python
import genomepy
genomepy.install_genome("hg38")
hg = genomepy.Genome("hg38")
```

```python
a = hg.get_seq("chr1",830000,832024)
b = hg.get_seq("chr1",860000,862024)
```

```python
dna = str(a).upper() + "[SEP]" + str(b).upper()
inputs = tokenizer(dna, return_tensors = 'pt')["input_ids"]
results = model(inputs.cuda())
hidden_states = results[0] # [1, sequence_length, 768]
```
